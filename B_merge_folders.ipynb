```python
import os 
import shutil
from tqdm import tqdm 

DESTINATION_NAME = 'All_fog_images'
FOLDERS = ['tiny_foggy', 'tiny_foggy_2']


EXPERIMENTS = ['RGB','RGB_distance']

DESTINATION = 'Radiate_Image_Dataset/' + DESTINATION_NAME

```

```python
try:
    des_list = os.listdir(DESTINATION)

except:
    os.mkdir(DESTINATION)
    os.mkdir(DESTINATION+"/coco_annotations")
    open(DESTINATION+"/coco_annotations/only_with_annotations.txt","w")
    os.mkdir(DESTINATION+"/fused_images")
    os.mkdir(DESTINATION+"/fused_images/RGB")
    os.mkdir(DESTINATION+"/fused_images/RGB_distance")
    os.mkdir(DESTINATION+"/fused_images/RGB_same")
    os.mkdir(DESTINATION+"/fused_images/RGB/train")
    os.mkdir(DESTINATION+"/fused_images/RGB/test")
    os.mkdir(DESTINATION+"/fused_images/RGB/valid")
    os.mkdir(DESTINATION+"/fused_images/RGB_distance/train")
    os.mkdir(DESTINATION+"/fused_images/RGB_distance/test")
    os.mkdir(DESTINATION+"/fused_images/RGB_distance/valid")
    os.mkdir(DESTINATION+"/fused_images/RGB_same/train")
    os.mkdir(DESTINATION+"/fused_images/RGB_same/test")
    os.mkdir(DESTINATION+"/fused_images/RGB_same/valid")
    
    des_list = os.listdir(DESTINATION)
```

```python
des_coco = open(DESTINATION+"/coco_annotations/only_with_annotations.txt", 'a')

for FOLDER in tqdm(FOLDERS):
    NAME = FOLDER
    FOLDER = "Radiate_Image_Dataset/"+FOLDER
    trg_coco = os.listdir(FOLDER+"/coco_annotations")
    

    for file in trg_coco:
        if file =="only_with_annotations.txt":
            with open(FOLDER+"/coco_annotations/"+file,"r") as ann:
                lines = ann.readlines()
                for line in lines:
                    line = NAME+"_"+line.strip()
                    des_coco.write(line+"\n")
            continue

        shutil.copyfile(FOLDER+"/coco_annotations/"+file,DESTINATION+"/coco_annotations/"+NAME+"_"+file)
    
    
    for EXPERIMENT in EXPERIMENTS:
        trg_train = os.listdir(FOLDER+"/fused_images/"+EXPERIMENT+"/train")
        trg_test = os.listdir(FOLDER+"/fused_images/"+EXPERIMENT+"/test")
        trg_valid = os.listdir(FOLDER+"/fused_images/"+EXPERIMENT+"/valid")

        for file in trg_train:
            shutil.copyfile(FOLDER+"/fused_images/"+EXPERIMENT+"/train/"+file,
                            DESTINATION+"/fused_images/"+EXPERIMENT+"/train/"+NAME+"_"+file)
        
        for file in trg_test:
            shutil.copyfile(FOLDER+"/fused_images/"+EXPERIMENT+"/test/"+file,
                            DESTINATION+"/fused_images/"+EXPERIMENT+"/test/"+NAME+"_"+file)
        
        for file in trg_valid:
            shutil.copyfile(FOLDER+"/fused_images/"+EXPERIMENT+"/valid/"+file,
                            DESTINATION+"/fused_images/"+EXPERIMENT+"/valid/"+NAME+"_"+file)


des_coco.close()
```

```python
import json

train_dic = {
    "info":
        {"year":"2023",
        "version":"1",
        "description":"Radiate Dataset",
        "contributor":"Heriot-Watt University",
        "url":"https://pro.hw.ac.uk/radiate/",
        "date_created":"2020-03-15"},
    "licenses":
        [{"id":1,
        "url":"https://choosealicense.com/licenses/mit/",
        "name":"MIT"}],
    "categories":[
        {"id":0,
        "name":"car",
        "supercategory":"none"},
        {"id":1,
        "name":"van",
        "supercategory":"none"},
        {"id":2,
        "name":"truck",
        "supercategory":"none"},
        {"id":3,
        "name":"bus",
        "supercategory":"none"},
        {"id":4,
        "name":"motorbike",
        "supercategory":"none"},
        {"id":5,
        "name":"pedestrian",
        "supercategory":"none"},
        {"id":6,
        "name":"group_of_pedestrians",
        "supercategory":"none"},
        {"id":7,
        "name":"bicycle",
        "supercategory":"none"}],
    "images":[],
    "annotations":[]}

test_dic = train_dic.copy()
valid_dic = train_dic.copy()

def create_json(folder,TARGET,img_len,ann_len):
    temp_img = []
    temp_ann = []
    for NAME in TARGET:
        with open(DESTINATION+"/fused_images/RGB/"+folder+"/"+NAME+"_"+"_annotations.coco.json",'r') as ann:
            dic = json.load(ann)
            for img in dic['images']:
                temp_img_dic = {
                        "id":len(temp_img)+img_len,
                        "license":1,
                        "file_name":NAME+"_"+img['file_name'],
                        "height":img['height'],
                        "width":img['width'],
                        "date_captured":"2020-03-15"}
                for ann in dic['annotations']:
                    if ann['image_id'] == img['id']:
                        temp_ann_dic = {
                        "id":len(temp_ann)+ann_len,
                        "image_id":len(temp_img)+img_len,
                        "category_id":ann['category_id'],
                        "bbox":ann['bbox'],
                        "area":ann['area'],
                        "segmentation":[],
                        "iscrowd":0}
                        temp_ann.append(temp_ann_dic)
                temp_img.append(temp_img_dic)
    return temp_img,temp_ann

train_dic['images'],train_dic['annotations'] = create_json("train",FOLDERS,0,0)

img_len = len(train_dic['images']) 
ann_len = len(train_dic['annotations'])

test_dic['images'],test_dic['annotations'] = create_json("test",FOLDERS,img_len,ann_len)

img_len += len(test_dic['images']) 
ann_len += len(test_dic['annotations'])

valid_dic['images'],valid_dic['annotations'] = create_json("valid",FOLDERS,img_len,ann_len)


train_json = json.dumps(train_dic, indent=5)
test_json = json.dumps(test_dic, indent=5)
valid_json = json.dumps(valid_dic, indent=5)

for EXPERIMENT in EXPERIMENTS:
    with open(DESTINATION+"/fused_images/"+EXPERIMENT+"/train/_annotations.coco.json", "w") as outfile:
        outfile.write(train_json)
    with open(DESTINATION+"/fused_images/"+EXPERIMENT+"/test/_annotations.coco.json", "w") as outfile:
        outfile.write(test_json)
    with open(DESTINATION+"/fused_images/"+EXPERIMENT+"/valid/_annotations.coco.json", "w") as outfile:
        outfile.write(valid_json)

for NAME in FOLDERS:
    for EXPERIMENT in EXPERIMENTS:
        os.remove(DESTINATION+"/fused_images/"+EXPERIMENT+"/train/"+NAME+"_"+"_annotations.coco.json")
        os.remove(DESTINATION+"/fused_images/"+EXPERIMENT+"/test/"+NAME+"_"+"_annotations.coco.json")
        os.remove(DESTINATION+"/fused_images/"+EXPERIMENT+"/valid/"+NAME+"_"+"_annotations.coco.json")
```
